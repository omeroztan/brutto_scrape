```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

df = pd.read_csv("results.csv")
df  # Just to look at the data
```

```python

data = []
for lap in df['Steuerklasse'].unique():
    data.append(
        go.Scatter(
            x=df[df['Steuerklasse']==lap]['Brutto Jahr'],
            y=df[df['Steuerklasse']==lap]['Netto Jahr'],
            name="Steuerklasse " + str(lap)
        )
    )

layout = go.Layout(
    title='Brutto - Netto',
    xaxis=dict(title='Brutto'),
    yaxis=dict(title='Netto'),
    autosize=False
)

fig = go.Figure(data=data, layout=layout)

max_range = max(df["Brutto Jahr"].max() + 20000, df["Netto Jahr"].max() + 20000)
fig.update_xaxes(range=[0, max_range], tick0=0, dtick=20000)
fig.update_yaxes(range=[0, max_range], tick0=0, dtick=20000)
fig.add_shape(
    type='line',
    x0=0, y0=0,  # Start at origin
    x1=max_range, y1=max_range,  # End at the max value of the range for both x and y
    line=dict(color='Gray', dash='dash'),
    xref='x', yref='y'  # Use 'x' and 'y' to refer to the axes
)

fig.update_yaxes(scaleanchor='x', scaleratio=1)
fig.update_layout(
    xaxis=dict(
        constrain='domain',  # This forces the aspect ratio to be maintained regardless of the plot size
    ),
    yaxis=dict(
        scaleanchor='x',
        scaleratio=1,
    ),
    width=800,
    height=600
)

fig.show()
```
